# The Fourth Order Langevin Dynamics Mean Vector Computation  

Consider the training data $Z=\{z_1,z_2,\cdots,z_n\}$, with $z_i=(X_i,y_i), i=1,2,\cdots,n$. Here, $X_i\in\mathbb{R}^{d}$ is a $d$-dimensional input and $y_i\in \mathbb{R}$ is the one-dimensional output. We consider a Ridge regression problem where the potential (loss) function is defined as  

\begin{equation}
    f(\theta) = \frac{1}{2n} \sum_{i=1}^{n} \left(y_i-\theta^TX_i\right)^2+\frac{\lambda}{2}\|\theta\|^2 = \frac{1}{2n}\|y-X\theta\|^2+\frac{\lambda}{2}\|\theta\|^2
\end{equation}  

The gradient of the loss is then given by  

\begin{align*}
    \nabla U(\theta) &= -\frac{1}{n} X^T(y-X\theta) +\lambda \theta \\
    &= \frac{1}{n}X^TX\theta -\frac{1}{n}X^Ty+\lambda\theta=\Big(\frac{1}{n}X^TX+\lambda I\Big)\theta-\frac{1}{n}X^Ty\\
\end{align*} 

Define $A=\Big(\frac{1}{n}X^TX+\lambda I\Big)$ and $b=\frac{1}{n}X^Ty$.  

Then $\nabla U(\theta)=A\theta-b$ is a linear function in $\theta$.  

## Mean Vector

The mean vector is defined as 

$$\begin{pmatrix}\theta \\ v_1 \\ v_2 \\ v_3\end{pmatrix}$$  

where, all the vectors are computed for given $\gamma$ and $\eta$ values and parameters $\mu_{ij}$ values and the calculations are as follows:  

### Component $\theta$:    

\begin{align*}
\theta & = - \int_{k\eta}^{(k+1)\eta} \int_{k\eta}^{s} \nabla U \Bigg(\theta + (r-k\eta)v_1 -\int_{k\eta}^{r}\int_{k\eta}^{w}\nabla U (\theta + (y-k\eta)v_1)dydw \\
& +\gamma v_2\frac{(r-k\eta)^2}{2!}+\gamma^2(-v_1+v_3)\frac{(r-k\eta)^3}{3!}\Bigg)drds\\
&+\gamma^2\int_{k\eta}^{(k+1)\eta}\int_{k\eta}^{s}\int_{k\eta}^{r}\int_{k\eta}^{w}\nabla U(\theta +(y-k\eta)v_1)dydwdrds\\
&+ \theta \mu_{00}+v_1\mu_{01}+v_2\mu_{02}+v_3\mu_{03}
\end{align*}

We use MATLAB for this computation. Let's split the integral into small parts as follows:

\begin{align*}
T_1 & = \int_{k\eta}^{r}\int_{k\eta}^{w}\nabla U (\theta + (y-k\eta)v_1)dydw\\
T_2 & = \gamma v_2\frac{(r-k\eta)^2}{2!}+\gamma^2(-v_1+v_3)\frac{(r-k\eta)^3}{3!}\\
T_3 & = \gamma^2\int_{k\eta}^{(k+1)\eta}\int_{k\eta}^{s}\int_{k\eta}^{r}\int_{k\eta}^{w}\nabla U(\theta +(y-k\eta)v_1)dydwdrds\\
T_4 & = \theta \mu_{00}+v_1\mu_{01}+v_2\mu_{02}+v_3\mu_{03}
\end{align*}

Then 

\begin{align*}
\theta & = - \int_{k\eta}^{(k+1)\eta} \int_{k\eta}^{s} \nabla U (\theta + (r-k\eta)v_1 - T_1 + T_2) drds + T_3 + T_4\\
\implies \theta & = \mu _{00}\,\theta +\mu _{01}\,v_{1}+\mu _{02}\,v_{2}+\mu _{03}\,v_{3}\\
& + \Big(\frac{\eta^4\gamma^2}{24}-\frac{\eta^2}{2}\Big)(A\theta-b)+\frac{\eta^4}{24}(A(A\theta-b))\\
& + \Big(\frac{\eta^5\gamma^2}{60}-\frac{\eta^3}{6}\Big)(Av_1)+\frac{\eta^5}{120}(A(Av_1))\\
& - \frac{\eta^4\gamma}{24}(Av_2)-\frac{\eta^5\gamma^2}{120}(Av_3)
\end{align*}  

Plugging $\mu_{00}=1$ and $\nabla U(\theta)=A\theta-b$ we get the simplified version of the parameter $\theta$ as follows:  

\begin{align*}
    \theta &=\theta + \mu_{01}v_1+\mu_{02}v_2+\mu_{03}v_3\\
    & + \left(\frac{\eta ^4\gamma ^2 }{24}-\frac{\eta^2}{2}\right) (A  \theta-b)+\frac{\eta^4}{24}(A(A\theta-b))\\
    &  + \left(\frac{ \eta ^5\gamma ^2}{60}-\frac{\eta ^3}{6}\right)A v_1+\frac{\eta ^5}{120} A (A v_1)\\
    & -\frac{\eta ^4\gamma}{24} A v_2 -\frac{\eta ^5\gamma ^2 }{120} A v_3
\end{align*}  

### Component $v_1$:

\begin{align*}
